```python
A = {1, 2, 5, 6}
B = {4, 2, 7, 10, 6}

def my_union(A, B):
    u = set()
    for a in A:
        u.add(a)
    for b in B:
        u.add(b)
    return u

def my_intersection(A, B):
    i = set()
    for a in A:
        if a in B:
            i.add(a)
    return i

def my_difference(A, B):
    d = A - my_intersection(A, B)
    return d

line1 = input().strip()
line2 = input().strip()

if line1 == "EMPTY":
    A = set()
else:
    A = set(map(int, line1.split()))
    
if line2 == "EMPTY":
    B = set()
else:
    B = set(map(int, line2.split()))

res = my_difference(A, B)

if not res:
    print("EMPTY")
else:
    print(" ".join(map(str, sorted(res))))

print(A.union(B), my_union(A, B))
print(A.intersection(B), my_intersection(A, B))
print(A.difference(B), my_difference(A, B))
```
